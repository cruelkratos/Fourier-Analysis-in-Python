```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap


dx= 0.001
L = np.pi
x =L* np.arange(-1,1,dx)
n = len(x) 
nquart = int(np.floor(n/4))
f = 0*x
f[nquart:2*nquart] = 4*np.arange(1,nquart+1)/n;
f[2*nquart:3*nquart] = 1-4*np.arange(0,nquart)/n;
plt.plot(x,f,'-',color ='b')
```

```python
#Approximating 
A= np.zeros(21)
B = np.zeros(21)
FS= 1/2* (np.sum(f*np.ones_like(x)))*dx
for i in range (21):
    A[i] = np.sum(f*np.cos(np.pi*(i+1)*x/L)) * dx
    B[i] = np.sum(f*np.sin(np.pi*(i+1)*x/L)) * dx
    FS = FS + A[i]*np.cos(np.pi*(i+1)*x/L) + B[i]*np.sin(np.pi*(i+1)*x/L)
    plt.plot(x,FS,'-',color='r')
plt.plot(x,f,'-',color='b')
```

Time Complexity is = $\mathcal{O}(n^2)$

```python
#Another Example
f2 = x**4 - 7*x**2 + 11
A= np.zeros(21)
B = np.zeros(21)
FS= 1/2* (np.sum(f*np.ones_like(x)))*dx
for i in range (21):
    A[i] = np.sum(f2*np.cos(np.pi*(i+1)*x/L)) * dx
    B[i] = np.sum(f2*np.sin(np.pi*(i+1)*x/L)) * dx
    FS = FS + A[i]*np.cos(np.pi*(i+1)*x/L) + B[i]*np.sin(np.pi*(i+1)*x/L)
    plt.plot(x,FS,'-',color='r')
plt.plot(x,f2,'-',color='b')
```
